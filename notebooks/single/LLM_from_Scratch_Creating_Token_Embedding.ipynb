```python
import torch
import torch.nn as nn
```

Token Id's for the words

```python
inp_ids = torch.tensor([2, 5, 4, 1])
```

Generating Embedding Weight Matrix

```python
vocab_size = 6
out_dims = 3

torch.manual_seed(123)
embedding_layer = nn.Embedding(vocab_size, out_dims)
```

```python
print(embedding_layer.weight)
```

```python
print(embedding_layer(torch.tensor([3])))
```

```python
print(embedding_layer(inp_ids))
```
