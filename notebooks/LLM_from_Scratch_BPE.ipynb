```python
import importlib
import tiktoken

print("TikToken version: ", importlib.metadata.version("tiktoken"))
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
text = (
    "Hello, Darsh this side.<|endoftext|> I was planning to connect with you"
    "Tell me where"
)
```

```python
integers = tokenizer.encode(text, allowed_special={"<|endoftext|>"})
print(integers)
```

```python
encoded_text = tokenizer.decode(integers)

print(encoded_text)
```
